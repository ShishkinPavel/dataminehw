# Data Quality Analysis

Анализ качества данных и стратегии очистки.

```python
import sys
sys.path.insert(0, '..')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from agents.data_quality_agent import DataQualityAgent
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('Set2')

df = pd.read_csv('../data/raw/dataset.csv')
agent = DataQualityAgent()
report = agent.detect_issues(df)

print(f"Shape: {df.shape}")
print(f"\nDtypes:\n{df.dtypes}")
print(f"\nDescribe:\n{df.describe()}")
df.head()
```

## 1.2 Пропущенные значения

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Heatmap пропусков
sns.heatmap(df.isnull(), cbar=True, yticklabels=False, cmap='RdYlGn_r', ax=axes[0])
axes[0].set_title('Heatmap пропущенных значений', fontsize=14)

# Bar chart
missing = report['missing']['by_column']
if missing:
    cols = list(missing.keys())
    vals = list(missing.values())
    pcts = [report['missing']['percent_by_column'][c] for c in cols]
    bars = axes[1].bar(cols, vals, color=sns.color_palette('Set2'))
    for bar, pct in zip(bars, pcts):
        axes[1].annotate(f'{int(bar.get_height())} ({pct}%)',
                         (bar.get_x() + bar.get_width() / 2., bar.get_height()),
                         ha='center', va='bottom', fontsize=11)
    axes[1].set_title('Пропуски по колонкам', fontsize=14)
    axes[1].set_ylabel('Количество')
else:
    axes[1].text(0.5, 0.5, 'Нет пропусков', ha='center', va='center', fontsize=14)

plt.tight_layout()
plt.show()

print(f"Всего пропусков: {report['missing']['total']}")
```

## 1.3 Дубликаты

```python
dup_info = report['duplicates']
print(f"Дубликатов: {dup_info['total']} ({dup_info['percent']}%)")

fig, ax = plt.subplots(figsize=(6, 6))
sizes = [len(df) - dup_info['total'], dup_info['total']]
labels_pie = [f"Уникальные ({sizes[0]})", f"Дубликаты ({sizes[1]})"]
ax.pie(sizes, labels=labels_pie, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'))
ax.set_title('Дубликаты vs уникальные записи', fontsize=14)
plt.tight_layout()
plt.show()

if dup_info['total'] > 0:
    print("\nПримеры дубликатов:")
    display(dup_info['examples'].head())
```

## 1.4 Выбросы

```python
numeric_cols = df.select_dtypes(include='number').columns.tolist()
n_cols = len(numeric_cols)

if n_cols > 0:
    fig, axes = plt.subplots(1, n_cols, figsize=(7 * n_cols, 6))
    if n_cols == 1:
        axes = [axes]
    for ax, col in zip(axes, numeric_cols):
        sns.boxplot(data=df, y=col, ax=ax, color=sns.color_palette('Set2')[0])
        ax.set_title(f'Boxplot: {col}', fontsize=12)
    plt.suptitle('Выбросы в числовых колонках', fontsize=14)
    plt.tight_layout()
    plt.show()

# Таблица выбросов
outlier_rows = []
for col, info in report['outliers']['by_column'].items():
    outlier_rows.append({
        'Колонка': col,
        'IQR count': info['iqr']['count'],
        'Z-score count': info['zscore']['count'],
        'Bounds (IQR)': f"{info['iqr']['bounds'][0]} — {info['iqr']['bounds'][1]}"
    })
pd.DataFrame(outlier_rows)
```

## 1.5 Дисбаланс классов

```python
imb = report['imbalance']
print(f"Ratio: {imb['ratio']}, Imbalanced: {imb['is_imbalanced']}")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Bar chart
dist = imb['distribution']
bars = axes[0].bar(dist.keys(), dist.values(), color=sns.color_palette('Set2'))
for bar in bars:
    axes[0].annotate(f'{int(bar.get_height())}',
                     (bar.get_x() + bar.get_width() / 2., bar.get_height()),
                     ha='center', va='bottom', fontsize=12)
axes[0].set_title('Распределение классов', fontsize=14)
axes[0].set_xlabel('Класс')
axes[0].set_ylabel('Количество')

# Grouped bar: label x source
if 'source' in df.columns:
    ct = df.groupby(['source', 'label']).size().reset_index(name='count')
    sns.barplot(data=ct, x='label', y='count', hue='source', ax=axes[1])
    axes[1].set_title('Классы по источникам', fontsize=14)

plt.tight_layout()
plt.show()
```

## Сводка обнаруженных проблем

Результаты анализа обнаружили:
- **Пропуски** в числовых колонках (text_length, rating)
- **Дубликаты** — ~3% строк
- **Выбросы** в text_length (значения > 5000 при среднем ~1000)
- **Дисбаланс классов** — quote-класс значительно меньше остальных


# Часть 2: Хирург — чистка данных

## 2.1 Стратегия A: conservative (median + clip_iqr)

```python
strategy_a = {'missing': 'median', 'duplicates': 'drop', 'outliers': 'clip_iqr'}
df_a = agent.fix(df, strategy=strategy_a)
comp_a = agent.compare(df, df_a)

print("Comparison: Original vs Strategy A")
display(comp_a)

# Boxplot после чистки
numeric_cols = df_a.select_dtypes(include='number').columns.tolist()
if numeric_cols:
    fig, axes = plt.subplots(1, len(numeric_cols), figsize=(7 * len(numeric_cols), 6))
    if len(numeric_cols) == 1:
        axes = [axes]
    for ax, col in zip(axes, numeric_cols):
        sns.boxplot(data=df_a, y=col, ax=ax, color=sns.color_palette('Set2')[1])
        ax.set_title(f'{col} (после Strategy A)', fontsize=12)
    plt.tight_layout()
    plt.show()
```

## 2.2 Стратегия B: aggressive (drop + clip_zscore)

```python
strategy_b = {'missing': 'drop', 'duplicates': 'drop', 'outliers': 'clip_zscore'}
df_b = agent.fix(df, strategy=strategy_b)
comp_b = agent.compare(df, df_b)

print("Comparison: Original vs Strategy B")
display(comp_b)

# Boxplot после чистки
numeric_cols = df_b.select_dtypes(include='number').columns.tolist()
if numeric_cols:
    fig, axes = plt.subplots(1, len(numeric_cols), figsize=(7 * len(numeric_cols), 6))
    if len(numeric_cols) == 1:
        axes = [axes]
    for ax, col in zip(axes, numeric_cols):
        sns.boxplot(data=df_b, y=col, ax=ax, color=sns.color_palette('Set2')[2])
        ax.set_title(f'{col} (после Strategy B)', fontsize=12)
    plt.tight_layout()
    plt.show()
```

## 2.3 Сравнительная таблица стратегий

```python
# Объединённая таблица
merged = comp_a[['metric', 'before']].copy()
merged.columns = ['metric', 'original']
merged['strategy_a'] = comp_a['after']
merged['strategy_b'] = comp_b['after']
display(merged)

# Grouped bar chart
fig, ax = plt.subplots(figsize=(12, 6))
x = range(len(merged))
width = 0.25
ax.bar([i - width for i in x], merged['original'], width, label='Original')
ax.bar(x, merged['strategy_a'], width, label='Strategy A')
ax.bar([i + width for i in x], merged['strategy_b'], width, label='Strategy B')
ax.set_xticks(x)
ax.set_xticklabels(merged['metric'], rotation=45, ha='right')
ax.set_title('Сравнение стратегий по метрикам', fontsize=14)
ax.legend()
plt.tight_layout()
plt.show()
```

# Часть 3: Обоснование выбора стратегии

## Контекст задачи
Задача — sentiment analysis текстовых отзывов. Ключевые требования:
- Сохранение максимального количества текстовых данных
- Корректная обработка числовых признаков (text_length, rating)
- Сохранение баланса классов

## Сравнение стратегий

**Strategy A (median + clip_iqr):**
- Сохраняет все строки (заполнение вместо удаления)
- IQR-клиппинг мягче: сохраняет больше вариативности
- Медиана устойчива к выбросам

**Strategy B (drop + clip_zscore):**
- Теряет строки при удалении пропусков (~10% данных)
- Z-score агрессивнее: предполагает нормальное распределение
- Чище с точки зрения "настоящих" данных

## Рекомендация
Для sentiment analysis рекомендуется **Strategy A**, потому что:
1. **Текстовые данные — главный признак**, терять строки нежелательно
2. **Числовые признаки (text_length, rating) вторичны** — медианное заполнение безопасно
3. **IQR не требует допущения о нормальности распределения**, что важно для text_length с мультимодальным распределением (разные источники)
4. Strategy A сохраняет 1010 строк vs 932 у Strategy B — на 8% больше обучающих данных
